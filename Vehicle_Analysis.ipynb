# Gender Analysis

```python
sns.countplot( x= 'Gender',color = 'pink', data =df)
plt.title("Gender Distribution")
plt.show()
```

***Observation: The dataset shows a slightly higher number of male customers compared to female customers.***

***Insight: The customer base is fairly balanced, which means insurance strategies should be designed to appeal to both genders equally.***

***Business Takeaway: Gender does not appear to be a strong differentiator in customer volume, so segmentation should rely more on age and interest levels.***

```python
sns.countplot(x='Gender', hue='Response', data=df)
plt.title("Gender vs Insurance Interest")
plt.show()
```

***Observation: Both genders show low overall interest in insurance, but males display slightly higher interest compared to females.***

***Insight: While the difference is small, it suggests that male customers may be marginally more receptive to insurance offerings.***

***Business Takeaway: Marketing campaigns could highlight gender‑neutral benefits, but a subtle emphasis on male‑oriented messaging might improve conversions.***

```python
sns.countplot(x = 'Vehicle_Age',color = 'green',data=df)
plt.title('Vehicle Age Analysis')
plt.show()
```

***Observation: Most vehicles are in the 1–2 year age range, followed by < 1 year, while the fewest are > 2 years old.***

***Insight: The customer base largely owns relatively new vehicles, which may influence their perception of insurance necessity.***

***Business Takeaway: Insurance products should emphasize long‑term protection for newer vehicles, while offering specialized plans for older vehicles to attract that smaller segment.***

```python
sns.countplot(x = 'Vehicle_Age',palette='dark:green',hue = 'Response',data=df)
plt.title('Vehicle_age v/s Response')
plt.show()
```

***Observation: Customers with older vehicles (> 2 years) show a higher proportion of positive insurance interest (Response = 1) compared to those with newer vehicles.***

***Insight: This suggests that risk perception increases with vehicle age, making older vehicle owners more receptive to insurance.***

***Business Takeaway: Marketing should highlight risk coverage and cost savings for older vehicles, while positioning insurance for newer vehicles as a future‑proof investment.***

# Vehicle Damage Analysis

```python
sns.countplot(x='Vehicle_Damage',color = 'orange', data=df)
plt.title('Vehicle_Damage Distribution')
plt.show()
```

***Observation: The dataset shows almost equal counts of vehicles with and without prior damage.***

***Insight: This balance indicates that vehicle damage is a common occurrence, making insurance relevant across the customer base.***

***Business Takeaway: Insurance providers should emphasize universal protection, since both damaged and undamaged vehicle owners represent significant segments.***

```python
sns.countplot(x='Vehicle_Damage', hue='Response', data=df)
plt.title("Vehicle Damage vs Insurance Interest")
plt.show()
```

***Observation: Customers with previously damaged vehicles are much more likely to show interest in insurance compared to those without damage.***

***Insight: Prior experience with damage increases risk awareness and perceived value of insurance.***

***Business Takeaway: Marketing campaigns should highlight real‑world scenarios of damage to build awareness among undamaged vehicle owners, while offering tailored plans for those with damage history.***

```python
sns.countplot(x='Previously_Insured', hue='Response', data=df)
plt.title("Previous Insurance vs Response")
plt.show()
```

***Observation: Customers who were not previously insured show higher interest in buying insurance compared to those already insured.***

***Insight: This suggests that new customers without prior coverage represent a strong target segment.***

***Business Takeaway: Marketing should focus on first‑time insurance buyers, emphasizing affordability and risk protection.***

```python
plt.figure(figsize = (8,4))
sns.histplot(df ['Annual_Premium'],bins =40)
plt.title('Annual Premium Distributions')
plt.show()
```

***Observation: Most premium values are concentrated below 100,000, with a few extreme outliers above 400,000.***

***Insight: The majority of customers prefer moderately priced insurance plans, while only a small group opts for very high premiums.***

***Business Takeaway: Insurance providers should design tiered premium structures — affordable plans for the majority, and premium packages for high‑value customers.***

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Vintage'], bins=30)
plt.title("Customer Vintage Distribution")
plt.show()
```

***Observation: Customer vintage (days since joining) is evenly spread across the dataset, with counts remaining fairly consistent across different ranges.***

***Insight: This indicates a balanced mix of new and long‑term customers, rather than a skew toward one group.***

***Business Takeaway:***

***For new customers, insurance companies can focus on onboarding offers and awareness campaigns.***

***For long‑term customers, strategies like loyalty rewards, premium upgrades, or retention discounts can strengthen engagement.***
